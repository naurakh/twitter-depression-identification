```python
import pandas as pd
import numpy as py
```

```python
dataset = pd.read_csv('dataset_labeled.csv')
dataset.head()
```

```python
dataset.drop(['Unnamed: 0'], axis=1)
```

```python
dataset['Text'] = dataset['Text'].str.lower()
print(dataset['Text'].head(5))
print('\n\n\n')
```

```python
import string 
import re #regex library

# import word_tokenize & FreqDist from NLTK
from nltk.tokenize import word_tokenize 
from nltk.probability import FreqDist
```

```python
def remove_tweet_special(text):
    # remove tab, new line, ans back slice
    text = text.replace('\\t'," ").replace('\\n'," ").replace('\\u'," ").replace('\\',"")
    # remove non ASCII (emoticon, chinese word, .etc)
    text = text.encode('ascii', 'replace').decode('ascii')
    # remove mention, link, hashtag
    text = ' '.join(re.sub("([@#][A-Za-z0-9]+)|(\w+:\/\/\S+)"," ", text).split())
    # remove incomplete URL
    return text.replace("http://", " ").replace("https://", " ")
                
dataset['Text'] = dataset['Text'].apply(remove_tweet_special)
```

```python
def remove_number(text):
    return  re.sub(r"\d+", "", text)

dataset['Text'] = dataset['Text'].apply(remove_number)
```

```python
def remove_punctuation(text):
    return text.translate(str.maketrans("","",string.punctuation))

dataset['Text'] = dataset['Text'].apply(remove_punctuation)
```

```python
def remove_whitespace_LT(text):
    return text.strip()

dataset['Text'] = dataset['Text'].apply(remove_whitespace_LT)
```

```python
def remove_whitespace_multiple(text):
    return re.sub('\s+',' ',text)

dataset['Text'] = dataset['Text'].apply(remove_whitespace_multiple)
```

```python
def remove_singl_char(text):
    return re.sub(r"\b[a-zA-Z]\b", "", text)

dataset['Text'] = dataset['Text'].apply(remove_singl_char)
```

```python
def word_tokenize_wrapper(text):
    return word_tokenize(text)

dataset['text_tokens'] = dataset['Text'].apply(word_tokenize_wrapper)

print('Tokenizing Result : \n') 
print(dataset['text_tokens'].head())
```

```python
def freqDist_wrapper(text):
    return FreqDist(text)

dataset['text_tokens_fdist'] = dataset['text_tokens'].apply(freqDist_wrapper)

print('Frequency Tokens : \n') 
print(dataset['text_tokens_fdist'].head().apply(lambda x : x.most_common()))
```

```python
from nltk.corpus import stopwords

# get stopword indonesia
list_stopwords = stopwords.words('indonesian')
```

```python
list_stopwords.extend(["yg", "dg", "rt", "dgn", "ny", "di", 'amp', 
                       'duh', 'nya', 'nih', 'sih','si', 'tau', 'tdk', 'tuh', 'utk', 'ya',
                       'sdh', 'aja', 'n', 't', 'wkwkwk',
                       'nyg', 'hehe', 'pen', 'u', 'nan', 'loh', 'rt',
                       '&amp', 'yah', 'dong', 'lah', 'eh', 'deh'])
```

```python
# read txt stopword using pandas
txt_stopword = pd.read_csv("stopwords.txt", names= ["stopwords"], header = None)

# convert stopword string to list & append additional stopword
list_stopwords.extend(txt_stopword["stopwords"][0].split(' '))
```

```python
list_stopwords = set(list_stopwords)
```

```python
def stopwords_removal(words):
    return [word for word in words if word not in list_stopwords]

dataset['text_tokens_rsw'] = dataset['text_tokens'].apply(stopwords_removal) 

print(dataset['text_tokens_rsw'].head())
```

```python
normalizad_word = pd.read_csv("normalisasi__fix_clean.csv")

normalizad_word_dict = {}
```

```python
for index, row in normalizad_word.iterrows():
    if row[0] not in normalizad_word_dict:
        normalizad_word_dict[row[0]] = row[1] 
```

```python
def normalized_term(document):
    return [normalizad_word_dict[term] if term in normalizad_word_dict else term for term in document]

dataset['text_normalized'] = dataset['text_tokens_rsw'].apply(normalized_term)

dataset['text_normalized'].head(10)
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import swifter
```

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
def stemmed_wrapper(term):
    return stemmer.stem(term)

term_dict = {}
```

```python
for document in dataset['text_normalized']:
    for term in document:
        if term not in term_dict:
            term_dict[term] = ' '
            
print(len(term_dict))
print("------------------------")
```

```python
for term in term_dict:
    term_dict[term] = stemmed_wrapper(term)
    print(term,":" ,term_dict[term])
    
print(term_dict)
print("------------------------")
```

```python
def get_stemmed_term(document):
    return [term_dict[term] for term in document]

dataset['text_tokens_stemmed'] = dataset['text_normalized'].swifter.apply(get_stemmed_term)
print(dataset['text_tokens_stemmed'])
```
